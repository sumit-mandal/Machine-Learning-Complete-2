## Cleaning our data

```python
plt.figure(figsize=(10,7))
sns.boxplot(x='Pclass',y='Age',data=df_train)
```

```python
## From graph it is clear that average age of Pclass 1 os 37 Pclass2 is 27 and Pclass3 is 24
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        
        if Pclass==1:
            return 37
        elif Pclass == 2:
            return 27
        else: 
            return 24
        
    else:
        return Age
```

```python
df_train['Age'] = df_train[['Age','Pclass']].apply(impute_age,axis=1)
```
