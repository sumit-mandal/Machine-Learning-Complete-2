```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0,5,11)
print(x)
y = x**2
```

```python
y
```

## Functional Plot for matplotlib

```python
plt.plot(x,y)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

## OOP method in matplotlib

```python
fig = plt.figure() #blank canvas

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left,right,width,height of the blank canvas

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_title('Title  sett')
```

```python
fig = plt.figure() #it creates a imaginery blamnk cannvas
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'g')
axes1.set_title('Larger plot',loc='left')
axes2.plot(x,y,'m')

axes2.set_title('Smaller plot')
```

# Matplotlib part 2

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)

```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First plot')
axes[1].plot(y,x)
axes[1].set_title('Second plot')
plt.tight_layout()
```

## Figure  Size and DPI

```python
fig = plt.figure(figsize=(3,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax = plt.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

# Part 3 Plot  Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green',linewidth=3,alpha=0.5,linestyle='dashdot',marker='o',markersize=10,
       markerfacecolor='yellow',markeredgecolor='red') #alpha shows transparency in line
```

## Plot range

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green',linewidth=3,alpha=0.5,linestyle='dashdot',marker='o',markersize=10,
       markerfacecolor='yellow',markeredgecolor='red') #alpha shows transparency in line

ax.set_xlim([0,2])
ax.set_ylim([0,2])
```

## Scatter plot

```python
plt.scatter(x,y)
```

## Histogram

```python
from random import sample
data = sample(range(1,1000),100)
plt.hist(data)
```

## BoxPlot

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(data,vert=True,patch_artist=True)
```
