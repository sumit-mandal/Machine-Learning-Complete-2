```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips
```

## lmplot()

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'])
```

```python
#scatter_kws changes size
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
          scatter_kws={'s':100})
```

```python
#seperating the cols
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
#seperating by rows and cols
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

```python
#seperating by rows and cols and hue
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time',hue='day')
```

## Changing size and aspect ratio

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time',hue='day',aspect=0.6,size=8)
```
