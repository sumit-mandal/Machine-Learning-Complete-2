```python
import numpy as np
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips
```

### barplot

```python
sns.barplot(x='sex',y='total_bill',data=tips) # it shows average of per categorical(m or f) value
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.jointplot(x='sex',y='total_bill',data = tips)
```

## countplot - visualising the count of columns. It is univariate

```python
sns.countplot(x='sex',data=tips)
```

## Boxplot

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

`Letâ€™s take the first box plot i.e, blue box plot of the figure and understand these statistical things:`
 - Bottom black horizontal line of blue box plot is minimum value
 - First black horizontal line of rectangle shape of blue box plot is First quartile or 25%
 - Second black horizontal line of rectangle shape of blue box plot is Second quartile or 50% or median.
 - Third black horizontal line of rectangle shape of blue box plot is third quartile or 75%
 - Top black horizontal line of rectangle shape of blue box plot is maximum value.
 - Small diamond shape of blue box plot is outlier data or erroneous data.

```python
sns.jointplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
# sns.set(rc={'figure.figsize':(80,30)})

```

## Violin plot

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

## StripPlot

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex')
```

## SwarmPlot- 
`It is similar to strip plot except the points are adjusted so that they font overlap`
`It is quite like combining a stripplot and violentplot`

```python

sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

## FactorPlot renamed to catplot

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
```
