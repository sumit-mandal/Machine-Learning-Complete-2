## `PCA is just a transformation of you data and attempts to find out what features explain the most variance in your data`

<img src= "PCA.png">

## Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing the data

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
cancer
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df
```

```python
print(cancer['target'])
print('name is' ,cancer['target_names'])
```

#### *We have 30 columns meaning 30-Dimensions for PCA. Maximum 3-D can be drawn*

## Scaling the components

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
scaled_data = sc.fit_transform(df)
print(scaled_data.shape)
scaled_data
```

## Performing PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2) #visualising entire 30-dimensional dataset just by 2 components
x_pca = pca.fit_transform(scaled_data)
```

```python
x_pca.shape # Dimensionality reduced from (569,30)---> (569,2)
```

## Visualising

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap='plasma')
plt.xlabel('First Principal Components')
plt.ylabel('Second Principal Components')
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_,columns=cancer['feature_names'])
```

```python
df_comp
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma')
```
