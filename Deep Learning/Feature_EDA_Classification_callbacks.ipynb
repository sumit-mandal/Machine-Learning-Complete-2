### Choosing values for X and Y

```python
X = df.drop('benign_0__mal_1',axis=1).values
y = df['benign_0__mal_1'].values
```

## Splitting the training and test data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=101)
```

```python
X_train
```

## Scaling the data

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

### Use keras model

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout

X_train.shape
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))
# model.add(Dense(30,activation='relu'))

#Binary Classification
model.add(Dense(1,activation='sigmoid'))

#Compiling 
model.compile(loss='binary_crossentropy',optimizer = 'adam')
```

### FItting the model

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test))
```

```python
model.save('Feature_EDA_Classification.h5')
```

### Using keras callbacks

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))
# model.add(Dense(30,activation='relu'))

#Binary Classification
model.add(Dense(1,activation='sigmoid'))

#Compiling 
model.compile(loss='binary_crossentropy',optimizer = 'adam')
```

```python
from keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor = 'val_loss',mode='min',patience=25,verbose=1)
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),callbacks=[early_stop],epochs=600)
```

```python
losses1 = pd.DataFrame(model.history.history)
losses1.columns=['Train_loss', 'Val_loss']
```

```python
losses1.plot()
```

### Use dropout layer to prevent overfitting

```python
from keras.layers import Dropout
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dropout(0.25))

model.add(Dense(15,activation='relu'))
model.add(Dropout(0.25))

# Binary Classification
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),callbacks=[early_stop])
```

```python
losses2 = pd.DataFrame(model.history.history)
losses2.columns=['Train_loss','Val_loss']
```

```python
losses2.plot()
```

### Predicting the classes

```python
predictions = model.predict_classes(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report,confusion_matrix

print('classification_report',classification_report(y_test,predictions))

print('confusion_matrix\n',confusion_matrix(y_test,predictions))
```
