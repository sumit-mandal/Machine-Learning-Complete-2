## Assigning values for X and y

```python
X = df.drop('price',axis=1).values
y = df['price'].values
y
```

```python
X
```

## Splitting test and train

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

## Scaling the value

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test) #we don't want to train on X_test hence not using fit_transfrom
```

### Creating the model

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
X_train.shape
```

## It looks like we have 19 incoming layer

```python
model = Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),batch_size=128,epochs=400,verbose=1)
```

## Making this history model as dataframe

```python
losses=pd.DataFrame(model.history.history)
```

```python
losses.plot()
```
